```python
%matplotlib inline

# import statements
import numpy as np
import matplotlib.pyplot as plt #for figures
import pandas as pd

MEarth = 5.972e24 #kge
R_Earth = 6371 #km
G = 6.67384e-11 #m3/kg/s2
```

```python
def Radius_planet(XFe, Mp):
    """ Radius of planet in km. 
    
    XFe is the weigth iron fraction, from 0.15 to 0.8 
    """ 
    return (7030-1840*XFe)*(Mp/MEarth)**0.282

def iron_mass_fraction_mantle(FeM):
    """ FeM is the iron number (iron mantle mole fraction), between 0 and 20. """
    M_Fe = 55.845
    M_Mg = 24.305
    M_Si = 28.0855
    M_O = 15.999
    M_tot = 2*((1-FeM/100)*M_Mg+FeM/100*M_Fe)+M_Si +4*M_O
    return 2*FeM/100.*M_Fe/M_tot

def core_mass_fraction(XFe, Mp, FeM):
    return (XFe-iron_mass_fraction_mantle(FeM))/1-iron_mass_fraction_mantle(FeM)

def Radius_core_hot(XFe, Mp, FeM): 
    """ Radius of core in km, hot case """
    X_CMF = core_mass_fraction(XFe, Mp, FeM)
    return 4850*X_CMF**0.328*(Mp/MEarth)**0.266

def Radius_core_cold(XFe, Mp, FeM): 
    """ Radius of core in km, cold case """
    X_CMF = core_mass_fraction(XFe, Mp, FeM)
    return 4790*X_CMF**0.328*(Mp/MEarth)**0.266

def Radius_core(XFe, Mp, FeM, case="hot"):
    if case == "hot":
        Rc = Radius_core_hot(XFe, Mp, FeM)
    else:
        Rc = Radius_core_cold(XFe, Mp, FeM)
    return Rc

def g0(XFe, Mp):
    """ surface gravity, units:m/s2 """
    return G*Mp/(Radius_planet(XFe, Mp)*1000)**2

def gCMB(XFe, Mp, FeM, case="hot"):
    """ surface gravity, units:m/s2 """
    Rc = Radius_core(XFe, Mp, FeM, case)
    return G*Mp*core_mass_fraction(XFe, Mp, FeM)/(Rc*1000)**2

def average_gravity_mantle(XFe, Mp, FeM, case="hot"):
    return (g0(XFe, Mp) + gCMB(XFe, Mp, FeM, case))/2

def average_density_mantle(XFe, Mp, FeM, case):
    V_mantle = 4/3*np.pi*(Radius_planet(XFe, Mp)**3-Radius_core(XFe, Mp, FeM, case)**3)*1000**3
    M_mantle = 1-core_mass_fraction(XFe, Mp, FeM)*Mp
    return M_mantle/V_mantle

def average_density_core(XFe, Mp, FeM, case="hot"):
    V_core = 4./3.*np.pi*(Radius_core(XFe, Mp, FeM, case)*1000)**3
    M_core = core_mass_fraction(XFe, Mp, FeM)*Mp
    return M_core/V_core

def Pcmb(XFe, Mp, FeM, case):
    """ Pressure at CMB in GPa """
    return average_gravity_mantle(XFe, Mp, FeM, case)*average_density_mantle(XFe, Mp, FeM, case)*\
                (Radius_planet(XFe, Mp)-Radius_core(XFe, Mp, FeM, case))*1000*1e-9

def alpha_mantle(XFe, Mp, FeM):
    return (13+0.738*core_mass_fraction(XFe, Mp, FeM)-11*(Mp/MEarth)**0.04)*1e-5

def Cp_mantle(FeM):
    return 1275.-585.*iron_mass_fraction_mantle(FeM)**1.06
    
def Tcmb_core(XFe, Mp, FeM, case):
    return 5400*(Pcmb(XFe, Mp, FeM, case)/140)**0.48/(1-np.log(1-FeM/100.))
    
def Tcmb_mantle(XFe, Mp, FeM, case):
    Tum = 2000 #K temperature in the uppermantle
    Dl = 500 #km thickness of the lithosphere
    inside_exp =  average_gravity_mantle(XFe, Mp, FeM, case)*alpha_mantle(XFe, Mp, FeM)/2/Cp_mantle(FeM)
    radius_lower_mantle = (Radius_planet(XFe, Mp) - Radius_core(XFe, Mp, FeM, case)-Dl)*1000
    return Tum*np.exp(inside_exp*radius_lower_mantle)
    
def TDsecond(XFe, Mp, FeM, case):
    Tum = 2000 #K temperature in the uppermantle
    Dl = 500 #km thickness of the lithosphere
    Dsecond = 500 #km thickness of D second
    inside_exp =  average_gravity_mantle(XFe, Mp, FeM, case)*alpha_mantle(XFe, Mp, FeM)/2/Cp_mantle(FeM)
    radius_lower_mantle = (Radius_planet(XFe, Mp) - Radius_core(XFe, Mp, FeM, case)-Dl-Dsecond)*1000
    return Tum*np.exp(inside_exp*radius_lower_mantle)

def DeltaT(XFe, Mp, FeM, case):
    return Tcmb_core(XFe, Mp, FeM, case) - TDsecond(XFe, Mp, FeM, case)
```

```python
import matplotlib.cm as cm


def make_planet(ax, XFe, Mp, FeM):

    R_p = Radius_planet(XFe, Mp)/(1.3*R_Earth)
    R_c = Radius_core_hot(XFe, Mp, FeM)/(1.3*R_Earth)
    rho_m = average_density_mantle(XFe, Mp, FeM, case="hot")
    rho_c = average_density_core(XFe, Mp, FeM, case="hot")
    
    circle_m = Circle((0, 0), R_p, facecolor='none',
                    edgecolor='k', linewidth=1, alpha=1)
    ax.add_patch(circle_m)
    circle_c = Circle((0, 0), R_c, facecolor='none',
                    edgecolor='k', linewidth=1, alpha=1)
    ax.add_patch(circle_c)

    im = ax.imshow(np.random.random((10, 10)),
                    origin='lower', cmap=cm.Greens,
                    interpolation='spline36',
                    extent=([-1, 1, -1, 1]))
    im.set_clip_path(circle_m)

    im = ax.imshow(np.random.random((50, 50)),
                    origin='lower', cmap=cm.magma,
                    interpolation='spline36',
                    extent=([-1, 1, -1, 1]))
    im.set_clip_path(circle_c)


    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])
    ax.set_aspect('equal', 'box')
    ax.axis("off")
    
    
    
```

```python
Mp_all= np.array([0.8, 1.0, 1.2, 1.4, 1.6, 1.8, 2.])*MEarth
XFe_all= [0.2, 0.4, 0.6, 0.8]
FeM_all = [0.,  0.2]

for FeM in FeM_all:
    
    fig, ax = plt.subplots(len(XFe_all), len(Mp_all), figsize=[14,8])
    
    for im, Mp in enumerate(Mp_all):
        for ix, XFe in enumerate(XFe_all):
            
            
            make_planet(ax[ix, im], XFe, Mp, FeM)
            
```

```python
import plotly.graph_objects as go
import numpy as np

# Create figure
fig = go.Figure()

# Add traces, one for each slider step
for step in np.arange(0, 5, 0.1):
    fig.add_trace(
        go.Scatter(
            visible=False,
            line=dict(color="#00CED1", width=6),
            name=" = " + str(step),
            x=np.arange(0, 10, 0.01),
            y=np.sin(step * np.arange(0, 10, 0.01))))

# Make 10th trace visible
fig.data[10].visible = True

# Create and add slider
steps = []
for i in range(len(fig.data)):
    step = dict(
        method="restyle",
        args=["visible", [False] * len(fig.data)],
    )
    step["args"][1][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active=10,
    currentvalue={"prefix": "Frequency: "},
    pad={"t": 50},
    steps=steps
)]

fig.update_layout(
    sliders=sliders
)

fig.show()
```
